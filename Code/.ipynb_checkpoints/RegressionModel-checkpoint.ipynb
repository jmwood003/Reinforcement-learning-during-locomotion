# Load the data

```python
os.chdir('C:\\Users\\Jonathan\\University of Delaware - o365\\Team-CHS-PT-Morton Lab - RL in Locomotion - Jonathan - RL in Locomotion - Jonathan\\Data')
D = sio.loadmat("regressVars.mat")
Iwsh = D['Iwsh'][:,0]
LrnVar = D['LrnVar'][:,0]
PrctSuc = D['PrctSuc'][:,0]
GID = D['GrpIdx'][0,:]-1 #0=RL group, 1=strat group
```


**Model:**
$$ Iwsh_i \approx Normal(\mu_i, \sigma) $$
$$ \mu_i \approx \alpha[Group] + \beta_{LrnVar}*LrnVar + \beta_{PrctSuc}*PrctSuc$$
$$ alpha \approx Normal(0,5) $$
$$ \beta_{LrnVar} \approx Normal(0,5) $$
$$ \beta_{PrctSuc} \approx Normal(0,5) $$
$$ \sigma \approx Exponential(1) $$

```python
#Build model using PM 
with pm.Model() as Iwsh_Model:
    alpha = pm.Normal("alpha", mu=0, sigma=5, shape=2)
    B_LV = pm.Normal("B_LV", mu=0, sigma=5)
    B_PC = pm.Normal("B_PC", mu=0, sigma=5)
    sigma = pm.Exponential("sigma", 1)    
    mu = pm.Deterministic("mu", alpha[GID] + B_LV * LrnVar + B_PC * PrctSuc)
    IW = pm.Normal("IW", mu=mu, sigma=sigma, observed=Iwsh)
    IWm_prior = pm.sample_prior_predictive()
    IWm_trace = pm.sample(1000, tune=1000)
```

```python
#Check the trace 
az.plot_trace(IWm_trace)
plt.show()
```

```python
#Check the prior
az.plot_density(
    [IWm_prior],
    group="prior",
    var_names=['alpha', "B_LV", "B_PC", "sigma"],
    point_estimate=None,
    )
plt.show()
```

```python
az.summary(IWm_trace, var_names=['alpha', "B_LV", "B_PC", "sigma"])
```
